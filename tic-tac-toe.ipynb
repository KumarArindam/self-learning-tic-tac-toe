```python
class tic_tac_toe():
    ## 1-x,0=0, 2=otherwise.
    
    def __init__(self,mode='training',ai_mode='ML'):
        self.board=np.fill((3,3),2)
        if mode == 'training':
            self.turn = 0  # 1:computer ,0: Human
        else:
            self.turn = 0
        
        self.game_finished=0
        self.moves=0    # number of moves played.
        self.generate_valid_moves()
        self.status=1   # 3;won, 2:Draw, 1:in progress, 0:lost
        self.ai_mode=ai_mode  # 'ML':next move based on trained algo, otherwise hard coded.
        
    def switch_turn(self):
        if self_turn == 1:
            self_turn = 0
        else:
            self_turn = 1
    
    def generate_valid_moves(self):
        # generate the list of all the possible moves.
        valid_moves_array=[]
        for i in range(3):
            for j in range(3):
                if self.board[i,j] == 2:
                    valid_moves_array.append(i,j)
                    
        self.valid_moves=valid_moves_array
    
    def update(self,position):
        # update status of game including board status, switch turns ,check for the result.
        if self.game_finished == 0:
            if(self.board[position]!=2): #fill only where there is a empty space, here 2=empty space.
                raise ValueError('invalid Move')
                
            # user playing
            self.board[position]=self.turn   # placing X(1) and 0(0)
            self.moves+=1
            self.switch_turn()
            # computer palying
            self.generate_valid_moves()
            self.update_status()          # checks for win/loss/draw/in progress and updates.
            
    def computer_basic_moves(self):

        for move in self.valid_moves:
        # loop through the rows to find a winning move.
            temp_board=self.board.copy()
            temp_board[move]=1
            
            for i in range(3):
                if list(temp_board[i,:]).count(1) == 3:
                    return move
                
        for move in self.valid_moves:
        #loop through the columns
            temp_board=self.board.copy()
            temp_board[move]=1
            for j in range(3):
                if list(temp_board[:,j]).count(1) == 3:
                    return move
                
        for move in self.valid_moves:
        #loop through one diagonal
            temp_board=self.board.copy()
            temp_board[move]=1
            if list(np.diag(temp_board)).count(1) == 3:
                return move
            
        
        for move in self.valid_moves:
        # loop through the other diagonal
            temp_board = self.board.copy()
            temp_board[move] = 1
            if list(np.diag(np.fliplr(temp_board))).count(1)==3:
                return move
            
        
        # finding the blocking moves
        for move in self.valid_moves:
            temp_board = self.board.copy()
            temp_board[move] = 0
            for i in range(3):
                if list(temp_board[i,:]).count(0) == 3:
                    return move
        
        for move in self.valid_moves:
            temp_board=self.board.copy()
            temp_board[move] = 0
            for j in range(3):
                if list(temp_board[:,j]).count(0) == 3:
                    return move
        
        for move in self.valid_moves:
            temp_board=self.board.copy()
            temp_board[move] = 0
            if list(np.diag(temp_board)).count(0) == 3:
                return move
        
        for move in self.valid_moves:
            temp_board=self.board.copy()
            temp_board[move]=0
            if list(np.diag(np.fliplr(temp_board))).count(0) == 3:
                return move
        
        ## if not wining or blocking then just play a random move.
        return random.choice(self.valid_moves)
                
    
    def play(self,posiion):
        #human followed by a computer move
        self.update(position)  ## players move
        
        if self.game_finished == 1:
            return None
        
        ## computer's next move
        if self.ai_mode == 'naive':
            computer_move = random.choice(self.valid_moves)
            self.update(computer_move)
            return computer_move
#         else if self.ai_mode == 'ML':
#             # computer playing based on the trained model
#             moves_rank={}
#             for move in self.valid_moves:
#                 create_input_array = np.array(list(self.board.flatten())+list(move)).reshape(1,-1)
#                 moves_rank[move] = model.predict(create_input_array)[0]
#             argmax=max(iter(moves_rank.keys()),key=(lambda key:moves_rank[key]))
            
#             self.update(argmax) # computer's move
#             return argmax
        
        else:
            computer_move = self.computer_basic_moves()
            self.update(computer_move)
            return computer_move
```
